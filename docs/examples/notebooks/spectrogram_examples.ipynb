# Spectrograms

Example of different types of spectrograms




```python
import numpy as np
import scipy as sp

from mth5.timeseries import spectrograms
from matplotlib import pyplot as plt
```

```python
n_samples = 4096

fs = 256  # Sampling frequency
dt = 1./fs
t = np.arange(n_samples) * dt

# Create a nonstationary signal: a sine wave with a frequency that changes over time
f0 = .1  # Initial frequency
f1 = 50  # Final frequency
signal = np.sin(2 * np.pi * (f0 + (f1 - f0) * t / max(t)) * t)

# an mt signal
b = []
for ff in np.logspace(-3, 3, 13):
    b.append(abs(np.log10(ff)) * np.sin(2*np.pi*t*ff+ np.pi/np.random.randint(1, 8)))

signal = np.array(b).sum(axis=0)
```

```python
plt.plot(t, signal)
```

```python
a = plt.specgram(signal, Fs=fs, noverlap=32)
plt.yscale("log")
plt.ylim(.001, 200)
```

```python
rtf, ftime, freq, stft = spectrograms.smethod(signal, df=fs)
```

```python
ax = plt.imshow(abs(rtf), aspect=20, origin="lower")
plt.yscale("log")
plt.ylim(.001, 200)
```
